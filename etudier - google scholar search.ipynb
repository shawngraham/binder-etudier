Ed Summer's [etudier](https://github.com/edsu/etudier) package.

Basic crawl, when you've got the google scholar page for a work:

```$ etudier.py 'https://scholar.google.com/scholar?start=0&hl=en&as_sdt=20000005&sciodt=0,21&cites=17950649785549691519&scipsc='```

Basic crawl, when you're searching for a particular term:

```
$etudier.py 'https://scholar.google.com/scholar?hl=en&as_sdt=0%2C21&q=cscw+memory&btnG='
```

for more pages of results, use `--pages`
for deeper results (eg citations of the citations, use `--depth`

This uses chromedriver, so periodically, you might have to prove that you're a human.

I'm writing this as a jupyter notebook in a virtual environment. I have not yet quite figured out how to make selenium and chromedriver work within mybinder.org.

````python
!etudier.py 'https://scholar.google.com/scholar?hl=en&as_sdt=0%2C21&q=disposition+of+human+remains&btnG='
````

[View disposition+of+human+remains.html](./disposition+of+human+remains.html)

You might want to open the .gexf in something like Gephi to prettify the output, etc.

If you see lots of little clumps, floating away from each other, I tend to interpret that as either search terms too broad, or a topic that does not cohere.
